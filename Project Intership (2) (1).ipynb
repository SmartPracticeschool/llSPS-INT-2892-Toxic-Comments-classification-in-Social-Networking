Stemming & Tokenization & Lemmatization

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
dt=[]
x=df['comment_text'].values
for i in range(0,4000):
    comment_text=re.sub('[^a-zA-Z]',' ',x[i])
    comment_text=comment_text.lower()
    comment_text=comment_text.split()
    comment_text=[ps.stem(word) for word in comment_text if not word in set(stopwords.words('english'))]
    comment_text= ' '.join(comment_text)
    dt.append(comment_text)
```

```python
dt
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 1000)
```

```python
z = cv.fit_transform(dt).toarray()
with open('CountVectorizer11','wb') as file:
    pickle.dump(cv,file)
```

```python
z.shape
```

```python
y=df.iloc[0:4000,3:4].values
y
```

Splitting the data into Test and Train

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(z,y,test_size=0.2,random_state=0)
```

ANN Model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(units=4000,init='random_uniform',activation='relu'))
model.add(Dense(units=3000,init='random_uniform',activation='relu'))
model.add(Dense(units=1,init='random_uniform',activation='sigmoid'))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

Optimizing Model

```python
model.fit(x_train,y_train,batch_size=32,epochs=3)
```

Saving the model

```python
model.save('toxic.h5')
```

```python
y_pred =  model.predict(x_test)
y_pred =  (y_pred>0.5)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
